### Program to find the first 20 non-even prime natural numbers.

```python
# Initialising variables and list
iCount = 0
iNaturalNumber = 1
lstOddPrimeNaturalNumbers = []

# Check whether the natural number is odd or not
def IsNumberOdd(iNumber):
    return ((iNumber%2) != 0)

# Check whether the natural number is non-even prime or not
def IsNumberPrime(iNumber):
    if IsNumberOdd(iNumber) and (iNumber>1):
        for iDivisor in range(2, iNumber):
            if (iNumber%iDivisor) == 0:
                return False
        return True
    else:
        return False
            
# Prepare list of first 20 non-even prime natural numbers
while iCount<20:
    if IsNumberPrime(iNaturalNumber):
        iCount = iCount + 1
        lstOddPrimeNaturalNumbers.append(iNaturalNumber)
    iNaturalNumber = iNaturalNumber + 1
    
# Print numbers from list
for i in lstOddPrimeNaturalNumbers:
    print(i)
```

### Program to implement 15 functions of string.

```python
# Initialise string
sString = input("Enter string:")

# Find length of the string
print("\n" + sString)
print("Length of the string:", len(sString), "\n")

# Initialise desired length
iDesiredLength = int(input("Enter the desired length of string:"))

# Right justify string
sRightJustified = sString.rjust(iDesiredLength)
print("String is right justified:")
print(sRightJustified + "\n")

# Strip whitespaces from left of the string
print("String is left stripped:", sRightJustified.lstrip())
print("\n")

# Left justify string
sLeftJustified = sString.ljust(iDesiredLength)
print("String is left justified:")
print(sLeftJustified + "." + "\n")

# Strip whitespaces from right of the string
print("String is right stripped:", sLeftJustified.rstrip() + ".")
print("\n")

# Center justify string
sCenterJustified = sString.center(iDesiredLength)
print("String is center justified:")
print(sCenterJustified + "." + "\n")

# Strip whitespaces from the string
print("String is stripped:", sCenterJustified.strip() + ".")
print("\n")

# Pad zeros at left of the string to make width 30
sFill = sString.zfill(iDesiredLength)
print("String is padded with zeros at left to make width:", len(sFill))
print(sFill + "\n")

# Swap cases for all the letters in the string
sCaseSwapped = sString.swapcase()
print("Case for all the letters in the string are swapped:")
print(sCaseSwapped + "\n")

# Convert string into list
lstWords = []
lstWords = sString.split(input("Enter delimiter:"))
print("Extracted words from the string into list:")
for s in lstWords:
    print(s + "\n")

# Join list with the separator
sStringJoined = input("Enter separator to join list:")
print("List joined:")
print(sStringJoined.join(lstWords) + "\n")

# Check whether all the characters in the string are lowercase
print("All letters in the string are lowercase:", sString.islower())
print("\n")

# Convert the string into lower case
print("Convert all the letters into lower case in the string:")
print(sString.lower() + "\n")

# Check whether all the characters in the string are uppercase
print("All letters in the string are uppercase:", sString.isupper())
print("\n")

# Convert the string into upper case
print("Convert all the letters into upper case in the string:")
print(sString.upper() + "\n")

# Convert all the first characters of the words in the string into upper case
print("First character of all the words in the string is made upper case:", sString.title())
print("\n")

# Check whether the characters in the string are alphanumeric
print("String is alphanumeric:", sString.isalnum())
print("\n")

# Check whether the characters in the string are only alphabets
print("String contains only alphabets:", sString.isalpha())
print("\n")
```

### Program to check if the given string is a Palindrome or Anagram or None of them. Display the message accordingly to the user.

```python
# Initialise string
sInputString = input("Enter a string:")
sAnagaramString = input("Enter a string to check for anagram of first string:")

# Check whether the string is palindrome or not
def IsPalindrome(sInput): 
    return (sInput == sInput[::-1])

# Check whether the string is anagram or not
def IsAnagram(sInput, sAnagarm):
    iIndex = 0
    lstInputString = list(sInput)
    lstAnagram     = list(sAnagarm)
    
    lstInputString.sort()
    lstAnagram.sort()

    while iIndex < len(sInput):
        if lstInputString[iIndex] == lstAnagram[iIndex]:
            iIndex = iIndex + 1
        else:
            return False

    return True

# Check for palindrome or anagram or none
sResultString = sInputString + " is"
if not IsPalindrome(sInputString.lower()):
    sResultString = sResultString + " not"
sResultString = sResultString + " a Palindrome \n"

sResultString = sResultString + sInputString + " and " + sAnagaramString + " are"
if not IsAnagram(sInputString.lower(), sAnagaramString.lower()):
    sResultString = sResultString + " not"
sResultString = sResultString + " Anagram."

# Printing result string
print(sResultString)
```

### Build a user defined function that removes all the additional characters from the string and converts it finally to lower case using built-in lower()

```python
# Convert string into list
sInputString = input("Enter a string:")

# Start of UDF
def GetAlphabeticString(sInput):
    # Initialising lists
    lstWords = list(sInput.lower())
    lstAlphabets = []
    
    # Initialising delimiter
    sDelimiter = ""
    
    for s in lstWords:
        if s.isalpha():
            lstAlphabets.append(s)
    return sDelimiter.join(lstAlphabets)
# End of UDF

# Printing output
print(GetAlphabeticString(sInputString))
```
